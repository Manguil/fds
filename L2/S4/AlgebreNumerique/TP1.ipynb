### Exercice 6

```python
def  vectNieme(n):
    v = np.zeros(n+1)
    v[0] = 1
    v[n] = -1
    return v
```

```python
n = int(input("n = "))
print(np.roots(vectNieme(n)))
```

### Exercice 7

```python
theta = np.pi/3
R = np.array([[np.cos(theta), 0, np.sin(theta)], [0, 1, 0], [-np.sin(theta), 0, np.cos(theta)]])
```

```python
Rsix = np.linalg.matrix_power(R,6)
print(Rsix)
I = np.eye(3)
print(np.allclose(I,Rsix))
```

### Exercice 10

```python
def longPol(n):
    racUnit = np.sort(np.roots(vectNieme(n)))
    perimetre = n * np.linalg.norm(racUnit[0] - racUnit[1])
    return perimetre

perCercle = 2*np.pi

max = 10
diff = []
for i in range(2,max):
    diff.append(perCercle - longPol(i))
plt.plot(diff,'.')
plt.show()
```
