### Exercice 6

```python
def vectNieme(n):
    v = np.zeros(n+1)
    v[0] = 1
    v[n] = -1
    return v
```

```python
n = int(input("n = "))
print(np.roots(vectNieme(n)))
```

### Exercice 7

```python
theta = np.pi/3
R = np.array([[np.cos(theta), 0, np.sin(theta)], [0, 1, 0], [-np.sin(theta), 0, np.cos(theta)]])
```

```python
Rsix = np.linalg.matrix_power(R,6)
print(Rsix)
I = np.eye(3)
print(np.allclose(I,Rsix))
```

### Exercice 8

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi)
y = x + np.sin(x)
plt.plot (x, y, 'g')
plt.title("courbe de y = x + sin(x)")
plt.show()
```

```python
plt.figure(1)
plt.plot(x,y, label = '1')
plt.figure(2)
plt.plot(x,x,label = '2')
plt.legend()
plt.show()
```
